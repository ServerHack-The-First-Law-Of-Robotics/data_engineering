```python
import json
import numpy as np
import pandas as pd

with open('../data/bonalog.json', encoding='utf-8') as file:
    nalog = json.load(file)
```

```python
len(nalog)
```

```python
extracted_nalog = {}

for inn, sample in nalog.items():
    # print("inn", inn)
    features = {}
    # latest_year = max(sample.keys())
    # print(latest_year)
    for year in sorted(sample.keys()):
        year_data = sample[year]
        # print(year_data["financialResult"])
        # features["capital_amount"] = year_data["financialResult"]["total3100"]
        if "balance" in year_data:
            if "current1700" in year_data["balance"]:
                features["actives"] = year_data["balance"]["current1700"]
            if "current2110" in year_data["financialResult"]:
                features["revenue"] = year_data["financialResult"]["current2110"]

        if "balance" in year_data:
            if "current1500" in year_data["balance"]:
                features["short_term_debt"] = year_data["balance"]["current1500"]
            
            if "current1400" in year_data["balance"]:
                features["long_term_debt"] = year_data["balance"]["current1400"]
    extracted_nalog[inn] = features
```
